# Univariate Graph Analysis


```python
# Univariate analysis
plt.figure(figsize=(10, 6))
sns.distplot(Total_rentalcount['cnt'], kde=True, bins=30)
plt.xlabel('cnt')
plt.ylabel('Count')
plt.title('Distribution of Total Number of Rentals')
plt.show()
```

```python
#Histogram
plt.figure(figsize=(10, 6))
sns.histplot(Total_rentalcount['cnt'], kde=True, bins=30)
plt.xlabel('cnt')
plt.ylabel('Count')
plt.title('Distribution of Total Number of Rentals')
plt.show()
```

```python
# Bivariate analysis
# Scatter plot
plt.scatter(Total_rentalcount['casual'], Total_rentalcount['registered'])
plt.title('Number of Registered Rentals vs. Number of Non-Registered Rentals')
plt.xlabel('registered')
plt.ylabel('casual')
plt.show()
```

```python
cross_tab = pd.crosstab(Total_rentalcount['holiday'], Total_rentalcount['weekday'])
cross_tab
```

```python
# Defining a function to separate categorical and numerical data 
def separate_categorical_and_continuous(Total_rentalcount):
    categorical_cols, numerical_cols = [] , []
    for col in Total_rentalcount.columns:
        if Total_rentalcount[col].dtypes == "object":
            categorical_cols.append(col)
        elif Total_rentalcount[col].dtypes in ["int64", "float64"]:
            numerical_cols.append(col)
    return categorical_cols, numerical_cols

categorical_cols, numerical_cols = separate_categorical_and_continuous(Total_rentalcount)
print("Categorical columns:", categorical_cols)
print("Numerical columns:", numerical_cols)
```

```python
# Calculating the average 'windspeed' for each unique 'atemp'
grouped_data = Total_rentalcount.groupby('atemp', as_index=False)['windspeed'].mean()
grouped_data
```

```python
# Calculating the average 'temp' for each unique 'weekday'
grouped_data1 = Total_rentalcount.groupby('weekday', as_index=False)['temp'].mean()
grouped_data1
```

```python
# Calculating and summarizing temperature-related statistics
grouped_data2 = Total_rentalcount.groupby('weekday').agg({'temp': ['mean', 'sum', 'count']})
grouped_data2
```

```python
# Calculating the sum of temperatures ('temp') for each unique combination of 'weekday' & 'weathersit'
grouped_data3 = Total_rentalcount.groupby(['weekday', 'weathersit'], as_index=False)['temp'].sum()
grouped_data3
```
